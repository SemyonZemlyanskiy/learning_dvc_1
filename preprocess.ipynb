```python
import pandas as pd
```

```python
df = pd.read_csv('data/diabet.csv')
df
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabet = load_diabetes(as_frame=True, scaled=False)
diabet
```

```python
df = pd.DataFrame(data = diabet['frame'])

df
```

```python
df.to_csv('data/diabet.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
y = df['target']
for x in diabet.feature_names:

    plt.scatter(df[x], y, color='blue', alpha=0.5)
    plt.xlabel('BMI')
    plt.ylabel('Disease Progression')
    plt.title(f'Зависимость прогрессирования диабета от {x}')
    plt.grid(True)
    plt.show()
    
```
